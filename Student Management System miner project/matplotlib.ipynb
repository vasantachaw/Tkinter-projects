```python
from matplotlib import pyplot as plt
import mysql.connector as mq
plt.figure(figsize=(10,5))


conn = mq.connect(host='localhost', user='root',
                          password='', database='dbsms')
cur = conn.cursor()
sql = 'select Grade,count(*) from marksheet where Grade=%s'
aPlush = ['A+']
a = ['A']
bPlush = ['B+']
b = ['B']
cPlush = ['C+']
c = ['C']
dPlush = ['D+']
d = ['D']
ePlush = ['E+']
e = ['E']
cur.execute(sql, aPlush)
aPlusha = cur.fetchone()
cur.execute(sql, bPlush)
bPlusha = cur.fetchone()
cur.execute(sql, cPlush)
cPlusha = cur.fetchone()
cur.execute(sql, dPlush)
dPlusha = cur.fetchone()
cur.execute(sql, ePlush)
ePlusha = cur.fetchone()

cur.execute(sql, a)
aa = cur.fetchone()
cur.execute(sql, b)
bb = cur.fetchone()
cur.execute(sql, c)
cc = cur.fetchone()
cur.execute(sql, d)
dd = cur.fetchone()
cur.execute(sql, e)
ee = cur.fetchone()

v=['A+','A','B+','B','C+','C','D+','D','E+','E']
x=[aPlusha[1],aa[1],bPlusha[1],bb[1],cPlusha[1],cc[1],dPlusha[1],dd[1],ePlusha[1],ee[1]]

#z=[33,55,44,67,88]
colors=['royalblue','pink','lightblue','thistle','orange','violet','lightgray','lightgreen']
plt.ylabel('Grade ',fontsize=16)
plt.bar(v,x,width=0.6,color=colors,linestyle='dotted',lw=2,label='SEM III',alpha=0.6)
#plt.title('RESULT',fontsize=16)
plt.legend(loc=2)
plt.savefig('images/graphs/grade.png',transparent=True)
plt.show()
```
